##create spark_session

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("SparkFirstExample.com").getOrCreate()
```

##creating dataframe using list of tuples

```python
data=[(1,"Dhanashri"),(2,"Digvijay"),(3,"Ratan")]
student_details_df=spark.createDataFrame(data,['id','name'])
student_details_df.show()
```

## creating DataFrame using list of lists

```python
My_data=[[4,"anil"],[5,"sunil"]]
student_details=spark.createDataFrame(My_data,['id','name'])
student_details.show()
```

## create DataFrame using list of dictionaries

```python
My_dict_data= [
   {'id': 6, 'name': 'suresh'},
   {'id': 7, 'name': 'ramesh'},
   {'id': 8, 'name': 'gyanesh'}
]

stud_dict_df=spark.createDataFrame(My_dict_data,['id','name'])
stud_dict_df.show()
```

```python
stud_dict_df.printSchema()
```

## how to define schema manually

```python
from pyspark.sql.types import *
schema = StructType([
    StructField('id', IntegerType(), True),
    StructField('name', StringType(), True)
])
df=spark.createDataFrame(My_dict_data,schema)
df.show()
df.printSchema()
```
